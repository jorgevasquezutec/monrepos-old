```python
from dbfread import DBF
import pandas as pd


TRASPASOS = 'Bases 2021/31-12-2021/Spm2003/2021/traspasos.DBF'
COSTOS = 'Bases 2021/31-12-2021/Spm2003/2021/costos22.DBF'
```

```python
#dbf_traspasos = DBF(TRASPASOS)
dbf_costos = DBF(COSTOS)
```

```python
#df_traspasos = pd.DataFrame(dbf_traspasos)
df_costos = pd.DataFrame(dbf_costos)
```

```python
df = df_costos
```

```python
# df = df[df]
#convert ccactual to in
df['CCACTUAL'] = df['CCACTUAL'].astype(int)
#select columns CCACTUAL, MODELO, NMODELO, COLOR, TALLA, CODIGO
df = df[['PEDIDO','ORDEN','CCACTUAL', 'MODELO', 'NMODELO', 'CODIGO']]
```

```python
 #Pivoteo del DataFrame para contar las cantidades por "modelo" y "talla"
# informe_talla = df.pivot_table(index=["MODELO","NMODELO","COLOR"],
#                                columns=["TALLA"], values="CODIGO", aggfunc="count", fill_value=0)

# Pivoteo del DataFrame para contar las cantidades por "modelo" y "ccactual"
informe_ccactual = df.pivot_table(index=["MODELO","NMODELO","PEDIDO",'ORDEN'],
                                  columns=["CCACTUAL"], values="CODIGO", aggfunc="count", fill_value=0)


# Calcular la columna "OTRAS" como la suma de ccactual diferentes de 100, 110 y 600
informe_ccactual['TEJIDO'] = informe_ccactual[100] + informe_ccactual[110] + informe_ccactual[112]
informe_ccactual['FINALIZADO'] = informe_ccactual[600]
informe_ccactual['EN_PROGRESO'] = informe_ccactual.sum(axis=1) - informe_ccactual['FINALIZADO'] - informe_ccactual['TEJIDO'] 

#SELECIONA LAS COLUMNAS 100, 110, 600, OTRAS_CCOSTO, MODELO, NMODELO, COLOR
# print(informe_ccactual.columns)
informe_ccactual = informe_ccactual[['TEJIDO','FINALIZADO','EN_PROGRESO']]
informe_ccactual = informe_ccactual.reset_index()
#print(informe_ccactual.columns)
#select columns MODEL,MODEL , TEJIDO, FINALIZADO, EN_PROGRESO
informe_ccactual = informe_ccactual[['PEDIDO','ORDEN','MODELO','NMODELO','TEJIDO','FINALIZADO','EN_PROGRESO']]
print(informe_ccactual)

# Unir los dos DataFrames por "modelo"
# informe = informe_talla.merge(informe_ccactual, on=["MODELO","NMODELO","COLOR"], how="left")

# Reiniciar el Ã­ndice
# informe.reset_index(inplace=True)

# print(informe)



#expor excel
informe_ccactual.to_excel('informe_ccactual.xlsx', index=True)

```
