```python
from dbfread import DBF
import pandas as pd
import dask.dataframe as dd
```

```python
""" 
    To do
    [ ] - Read All DBF files Traspasos
    [ ] - Combine all DBF files Traspasos
    [ ] - Read All DBF files costos22
    [ ] - Combine all DBF files costos22
"""
```

```python

COLUMNAS_TRASPASOS = ['NROTRASP','FECHORA']
COLUMNAS_COSTOS22 = [
    "CODCLI",
    "CLIENTE",
    "CANTIDAD",
    "INICIO",
    "ENTREGA",
    "MODELO",
    "NMODELO",
    "COLOR",
    "TALLA",
    "CODIGO",
    "PRENDA",
    "NPRENDA",
]
for i in range(1, 25):
    COLUMNAS_COSTOS22.append(f"PASO{i}")
    COLUMNAS_COSTOS22.append(f"VEZ{i}")
    COLUMNAS_COSTOS22.append(f"CCOSTO{i}")
    COLUMNAS_COSTOS22.append(f"NCCOSTO{i}")
    COLUMNAS_COSTOS22.append(f"FECHA{i}")

```

```python
years = [2021,2022]
df_traspasos = dd.from_pandas(pd.DataFrame(), npartitions=3)
df_costos = dd.from_pandas(pd.DataFrame(), npartitions=3)
for year in years:
    path = f'Bases {year}/31-12-{year}/Spm2003/{year}/'
    traspasos_path = path + 'traspasos.DBF'
    costos_path = path + 'costos22.DBF'
    #traspasos
    dbf = DBF(traspasos_path)
    df = pd.DataFrame(iter(dbf))
    df = df[COLUMNAS_TRASPASOS]
    ddf = dd.from_pandas(df, npartitions=3)
    dff = ddf.astype({'NROTRASP': 'int64'})
    df_traspasos = dd.concat([df_traspasos,ddf], axis=0)
    #costos
    dbfc = DBF(costos_path)
    dfc = pd.DataFrame(iter(dbfc))
    dfc = dfc[COLUMNAS_COSTOS22]
    for i in range(1, 25):
        dfc[f"PASO{i}"] = dfc[f"PASO{i}"].fillna(0).astype('int64')
        dfc[f"FECHA{i}"] = dfc[f"FECHA{i}"].astype('datetime64[ns]')
    ddfc = dd.from_pandas(dfc, npartitions=3)
    df_costos = dd.concat([df_costos,ddfc], axis=0)
    
```

```python
df_costosf = df_costos
```

```python
df_costosf
```

```python
columns = [f'PASO{i}' for i in range(1, 25)]

for idx,column in enumerate(columns):
    df_costosf = df_costosf.merge(df_traspasos, left_on=column, right_on='NROTRASP', how='left', suffixes=('',f'_{column}'))

df_costosf = df_costosf.rename(columns={'FECHORA': 'FECHORA_PASO1'})
df_costosf = df_costosf.rename(columns={'NROTRASP': 'NROTRASP_PASO1'})

# for idx,column in enumerate(columns):
#     df_costosf = df_costosf.drop(columns=[f'NROTRASP_{column}'])
#     df_costosf[f'FECHA{idx}'] = df_costosf[f'FECHORA_{column}']
#     df_costosf = df_costosf.drop(columns=[f'FECHORA_{column}'])
```
